```
import numpy as np
```

### Julia's statistic

```
## Julia's statistic

def dist(x,y):
  n = x.shape[0]
  t = (np.abs(x - y.mean()) <= np.abs(x - x.mean())).astype(int)
  t +=  (np.abs(y - x.mean()) <= np.abs(y - y.mean())).astype(int)
  return t.sum()/2/n

n = 100000
statistic = np.zeros(n)
for k in range(n):
  x, y = np.random.randn(4), np.random.randn(4)
  statistic[k] = dist(x,y)

print('Mean of test statistic: {:.4f}'.format(statistic.mean()))
```

```
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt

plt.hist(statistic,bins=100)
plt.show()
```

### Alternative

```
def kNN(x,y):
  def kNN_indicator(xtest,x,y):
    if min(np.abs(xtest-x)) < min(np.abs(xtest-y)):
      return 1
    else:
      return 0
  
  n = x.shape[0]
  m = y.shape[0]
  T = 0 
  for k in range(n):
    T += kNN_indicator(x[k],np.delete(x,k),y)
  for k in range(m):
    T += kNN_indicator(y[k],np.delete(y,k),x)
  
  return T/(n+m)

n = 100000
N = 4
statistic = np.zeros(n)
for k in range(n):
  x, y = np.random.randn(N), np.random.randn(N)
  statistic[k] = kNN(x,y)

statistic = statistic*(2*N-1)/2/(N-1)
print('Mean of test statistic: {:.4f}'.format(statistic.mean()))


```

```
plt.hist(statistic,bins=100)
plt.show()
```
